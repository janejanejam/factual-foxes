```python
# Import libraries
import pandas as pd     
import numpy as np
from pytrends.request import TrendReq
```

```python
csv = pd.read_csv('webscrape.csv')
df = pd.DataFrame(csv)
df = df.drop(['Unnamed: 0'], axis=1).rename(columns={'label text':'label_text'})
df
```

```python
kw_list = ['george floyd', 'ahmaud arbery', 'murder hornet', 'spacex launch', 'derek chauvin']
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
pytrends.build_payload(kw_list, cat=0, timeframe = '2020-05-01 2020-05-31', geo='US', gprop='')
```

```python
df_query=pytrends.interest_over_time()
df_query=df_query.drop(['isPartial'], axis =1)
df_query
```

```python
averageList = []
for item in kw_list:
    averageList.append(df_query[item].mean())
```
